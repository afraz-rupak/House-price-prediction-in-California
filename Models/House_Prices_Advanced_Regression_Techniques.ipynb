```python
train_path='/content/drive/MyDrive/Data SET/HousePrices_train.csv'
test_path='/content/drive/MyDrive/Data SET/HousePrices_test.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
train = train.drop('Id', axis=1)
test = test.drop('Id', axis=1)
```

```python
#LotFrontage    1201 non-null   float64
LotFrontage_missing_index = train[train['LotFrontage'].isnull()].index.tolist()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer


mice_imputer = IterativeImputer()
mice_imputer = mice_imputer.fit(train[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']])

train[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']] = mice_imputer.transform(train[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']])
train.iloc[LotFrontage_missing_index ]
```

```python
#LotFrontage    1201 non-null   float64
LotFrontage_missing_index = test[test['LotFrontage'].isnull()].index.tolist()

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer


mice_imputer = IterativeImputer()
mice_imputer = mice_imputer.fit(test[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']])

test[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']] = mice_imputer.transform(test[['LotFrontage','LotArea','OverallQual','TotalBsmtSF','1stFlrSF','GrLivArea','TotRmsAbvGrd','GarageArea']])
test.iloc[LotFrontage_missing_index ]
```

```python
from sklearn.impute import SimpleImputer
for col in train.columns:
    if train[col].dtype in [np.float64, np.int64]:
        sip = SimpleImputer(missing_values=np.nan, strategy='mean')
        train[col] = sip.fit_transform(train[col].values.reshape(-1, 1))[:, 0]
    else:
        sip = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value="Not_specified")
        train[col] = sip.fit_transform(train[col].values.reshape(-1, 1))[:, 0]

for col in test.columns:
    if test[col].dtype in [np.float64, np.int64]:
        sip = SimpleImputer(missing_values=np.nan, strategy='mean')
        test[col] = sip.fit_transform(test[col].values.reshape(-1, 1))[:, 0]
    else:
        sip = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value="Not_specified")
        test[col] = sip.fit_transform(test[col].values.reshape(-1, 1))[:, 0]
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
train.info()
```

```python
obj=[]
for i in range(len(train.columns)):
  t = train.columns[i]
  if train[t].dtypes == 'object':
    obj.append(t)
```

```python
for i in range(len(obj)):
  p = obj[i]
  train[p]= label_encoder.fit_transform(train[p])
```

```python
obj=[]
for i in range(len(test.columns)):
  t = test.columns[i]
  if test[t].dtypes == 'object':
    obj.append(t)

for i in range(len(obj)):
  p = obj[i]
  test[p]= label_encoder.fit_transform(test[p])
```

```python
x = train.iloc[:, :-1]
y = train.iloc[:, -1]
x_test = test
```

```python
x.shape
```

**Feature Engineering**

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func = f_classif)
fit_features.fit(x,y)
fs = pd.DataFrame(fit_features.scores_,index=x.columns, columns = ['score values'])
fs.nlargest(50,'score values')
```

```python
col = []
for i in range(len(fs)):
  if fs.values[i] >= 2:
    col.append(fs.index[i])
```

```python
len(col)
```

```python
x_new = x[col]
```

```python
x_test_main = test[col]
```

```python
x_test_main.shape,x_new.shape,y.shape
```

Train/Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size=0.2)
```

**Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
lgrg = LogisticRegression()
lgrg.fit(x_train,y_train)
lgrg.score(x_test,y_test)*100
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 20)
rf.fit(x_train, y_train) 
rf.score(x_test,y_test)*100
```

Ridge
